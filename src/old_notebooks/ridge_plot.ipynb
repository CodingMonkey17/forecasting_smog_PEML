```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_PM25_2017_tidy = pd.read_csv("../data/data_tidy/2017_PM25_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_PM10_2017_tidy = pd.read_csv("../data/data_tidy/2017_PM10_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_O3_2017_tidy = pd.read_csv("../data/data_tidy/2017_O3_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_NO2_2017_tidy = pd.read_csv("../data/data_tidy/2017_NO2_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_PM25_2018_tidy = pd.read_csv("../data/data_tidy/2018_PM25_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_PM10_2018_tidy = pd.read_csv("../data/data_tidy/2018_PM10_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_O3_2018_tidy = pd.read_csv("../data/data_tidy/2018_O3_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_NO2_2018_tidy = pd.read_csv("../data/data_tidy/2018_NO2_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_NO2_2019_tidy = pd.read_csv("../data/data_tidy/2019_NO2_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_PM25_2020_tidy = pd.read_csv("../data/data_tidy/2020_PM25_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_PM10_2020_tidy = pd.read_csv("../data/data_tidy/2020_PM10_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_O3_2020_tidy = pd.read_csv("../data/data_tidy/2020_O3_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_NO2_2020_tidy = pd.read_csv("../data/data_tidy/2020_NO2_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_PM25_2021_tidy = pd.read_csv("../data/data_tidy/2021_PM25_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_PM10_2021_tidy = pd.read_csv("../data/data_tidy/2021_PM10_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_O3_2021_tidy = pd.read_csv("../data/data_tidy/2021_O3_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_NO2_2021_tidy = pd.read_csv("../data/data_tidy/2021_NO2_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_PM25_2022_tidy = pd.read_csv("../data/data_tidy/2022_PM25_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_PM10_2022_tidy = pd.read_csv("../data/data_tidy/2022_PM10_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_O3_2022_tidy = pd.read_csv("../data/data_tidy/2022_O3_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_NO2_2022_tidy = pd.read_csv("../data/data_tidy/2022_NO2_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_PM25_2023_tidy = pd.read_csv("../data/data_tidy/2023_PM25_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_PM10_2023_tidy = pd.read_csv("../data/data_tidy/2023_PM10_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_O3_2023_tidy = pd.read_csv("../data/data_tidy/2023_O3_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
df_NO2_2023_tidy = pd.read_csv("../data/data_tidy/2023_NO2_tidy.csv", index_col=0, sep=';', decimal='.', encoding='utf-8')
```

```python
frames_2017 = [df_NO2_2017_tidy, df_O3_2017_tidy, df_PM10_2017_tidy, df_PM25_2017_tidy]
frames_2018 = [df_NO2_2018_tidy, df_O3_2018_tidy, df_PM10_2018_tidy, df_PM25_2018_tidy]
frames_2020 = [df_NO2_2020_tidy, df_O3_2020_tidy, df_PM10_2020_tidy, df_PM25_2020_tidy]
frames_2021 = [df_NO2_2021_tidy, df_O3_2021_tidy, df_PM10_2021_tidy, df_PM25_2021_tidy]
frames_2022 = [df_NO2_2022_tidy, df_O3_2022_tidy, df_PM10_2022_tidy, df_PM25_2022_tidy]
frames_2023 = [df_NO2_2023_tidy, df_O3_2023_tidy, df_PM10_2023_tidy, df_PM25_2023_tidy]
```

```python
for df in frames_2017:
    df['year'] = '2017'
for df in frames_2018:
    df['year'] = '2018'
for df in frames_2020:
    df['year'] = '2020'
for df in frames_2021:
    df['year'] = '2021'
for df in frames_2022:
    df['year'] = '2022'
for df in frames_2023:
    df['year'] = '2023'
```

```python
df_NO2 = pd.concat([
    frames_2017[0],
    frames_2018[0],
    frames_2020[0],
    frames_2021[0],
    frames_2022[0],
    frames_2023[0]
])
df_O3 = pd.concat([
    frames_2017[1],
    frames_2018[1],
    frames_2020[1],
    frames_2021[1],
    frames_2022[1],
    frames_2023[1]
])
df_PM10 = pd.concat([
    frames_2017[2],
    frames_2018[2],
    frames_2020[2],
    frames_2021[2],
    frames_2022[2],
    frames_2023[2]
])
df_PM25 = pd.concat([
    frames_2017[3],
    frames_2018[3],
    frames_2020[3],
    frames_2021[3],
    frames_2022[3],
    frames_2023[3]
])
```

```python
print(df_NO2.shape)
print(df_O3.shape)
print(df_PM10.shape)
print(df_PM25.shape)
```

Now, plot:

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Computer Modern Serif",
    "axes.labelsize": 8,     # fontsize for x and y labels
    "xtick.labelsize": 8,    # fontsize of the tick labels
    "ytick.labelsize": 8,    # fontsize of the tick labels
    "legend.fontsize": 8,    # fontsize of the legen
})
```

```python
# https://seaborn.pydata.org/examples/kde_ridgeplot.html
# is the source of the plotting code below

def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, r'$\textbf{\textrm{' + label + '}}$', color = color,
            ha = "left", va = "center", transform = ax.transAxes)
```

```python
pal = sns.dark_palette("#000000", n_colors = 10, reverse = True)
g = sns.FacetGrid(df_NO2, row = "year", hue = "year",
                  aspect = 12,
                  height = .5, palette = pal)
g.map(sns.kdeplot, "NL10636",
      bw_adjust = .5, clip_on = False,
      fill = True, alpha = 1, linewidth = 1.5)
g.map(sns.kdeplot, "NL10636", clip_on = False, color = "w", lw = 2, bw_adjust = .5)
g.refline(y = 0, linewidth = 2, linestyle = "-", color = None, clip_on = False)
g.map(label, "NL10636")
g.figure.subplots_adjust(hspace = -.25)
g.set_titles("")
g.set(yticks = [], ylabel = "")
# plt.title(r'$\textrm{NO}_{2}$', loc = 'center', y = 4.4)
plt.xlabel('')
g.despine(bottom = True, left = True)

plt.savefig("NO2_ridge_distribution.pdf",
            format = 'pdf',
            bbox_inches = 'tight',
            pad_inches = 0.05)
```
